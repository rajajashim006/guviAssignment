```python
from tkinter import *
from tkinter import messagebox
import os
import re

root = Tk()
root.geometry("800x150")
root.title("Welcome to guvi")
t= Label(root, text = "Welcome to Guvi", font = ('arial',15))
t.pack()

un = StringVar()
pw = StringVar()

#file to store the data guvi_Assignment_Details.txt
def rg():
    un1 = t1.get()
    pw1 = t2.get()
    
    try:
        str1 = un1
        rp1 ="[A-Za-z]+@+[A-Za-z]+\.+[A-Za-z]+"
        rc1 = re.compile(rp1)
        results = rc1.match(str1)
        #print(results)
        un2 = results.group()

        str2 = pw1
        rp2 = "[A-Za-z0-9 !@.\/_]{6,20}"
        rc1 = re.compile(rp2)
        results1 = rc1.match(str2)
        #print(results1)
        pw2 = results1.group()

        f = open("guvi_Assignment_Details.txt","a")
        f.write(results.group() +'\n')
        f.write(results1.group() +'\n')
        f.close()
    except AttributeError:
        messagebox.showinfo(title="Please follow the standard while creating the Username and Password", message="1. Email address should contain <alphabets>@<domain.com> \n 2. Password should have 1 [Upper, Lower, Special character, Number] \n 3. Password should be between 6 to 16 characters")

def lg():
    #f1=open("guvi_Assignment_Details.txt","r")
    un2=t1.get()
    pw2=t2.get()
    
    with open("guvi_Assignment_Details.txt","r") as fp:
        rs=fp.readlines()
        for row in rs:
            if (row.find(un2)!= -1) and (row.find(pw2)!= -1):
                messagebox.showinfo(title="Guvi", message="Entered Username and Password not matching")
                break
            else:
                messagebox.showinfo(title="Guvi", message="Login Successful")
                break
        fp.close()
        
form = Frame(root)
form.pack(side= TOP)


l1 = Label(form, text = "Username", font = ('arial', 12))
l2 = Label(form, text = "Password", font = ('arial', 12))
l1.grid(row=1, sticky= W)
l2.grid(row=2, sticky= W)

t1= Entry(form, textvariable = un)
t2= Entry(form, textvariable = pw, show = "*")
t1.grid(row= 1, column= 2)
t2.grid(row= 2, column = 2)

b1 = Button(form, text ="Register",command = rg)
b2 = Button(form, text = "Login", command = lg)
b1.grid(row= 8, column = 1)
b2.grid(row= 8, column= 2)



mainloop()
```
